```python
# O intuito desse projeto é criar um calcculador de score de crédito de clientes utilizando as bases de dados fornecidas
# Os scores vão ser classificados como Bom, Ok ou Ruim
# Importação de bibliotecas

import pandas as pd # Para manipulação de dados
from sklearn.preprocessing import LabelEncoder # Para transformar variáveis categóricas em numéricas
from sklearn.model_selection import train_test_split # Para dividir os dados em treino e teste
```

```python
# Importar base de dados
df = pd.read_csv("clientes.csv")
display(df.head(10))
df.info()
```

```python
# Tratar e preparar base para IA
# As colunas profissao, mix_credito, comportamento_pagamento, score_credito, são strings
# IA só consegue ler números, então precisamos transformar as colunas em numericas
# Label encoder transforma cada texto em número, por exemplo: "Bom" vira 0, "Ok" vira 1 e "Ruim" vira 2

codificador_profissao = LabelEncoder()
df["profissao"] = codificador_profissao.fit_transform(df["profissao"])

codificador_mix_credito = LabelEncoder()
df["mix_credito"] = codificador_mix_credito.fit_transform(df["mix_credito"])

codificador_comportamento_pagamento = LabelEncoder()
df["comportamento_pagamento"] = codificador_comportamento_pagamento.fit_transform(df["comportamento_pagamento"])
```

```python
# Como funciona  a criação de uma IA?
# Vou ter duas variáveis, x e y
# Y é a variável que eu quero prever, no caso o score de crédito
# X são as variáveis que eu vou usar para prever o score de crédito

y = df["score_credito"]

x = df.drop(columns=["score_credito", "id_cliente"])

# Train test split é uma função que divide os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# Criar o modelo de previsão
# Vou usar a Árvore de Decisão e o Nearest Neighbors

# Primeiro importamos a IA
    
from sklearn.ensemble import RandomForestClassifier # Árvore de Decisão
from sklearn.neighbors import KNeighborsClassifier # Nearest Neighbors
    
# Então criamos a IA
arvore_decisao = RandomForestClassifier()
vizinhos_proximos = KNeighborsClassifier()

# E finalmente treinamos a IA
arvore_decisao.fit(x_treino, y_treino)
vizinhos_proximos.fit(x_treino, y_treino)
```

```python
# Avaliar e escolher melhores modelos

# Fazer a previsão
previsao_arvore_decisao = arvore_decisao.predict(x_teste)
previsao_vizinhos_proximos = vizinhos_proximos.predict(x_teste)

# Comparar as previsões e escolher o melhor modelo
from sklearn.metrics import accuracy_score # Para comparar as previsões

print(accuracy_score(y_teste, previsao_arvore_decisao))
print(accuracy_score(y_teste, previsao_vizinhos_proximos))

# Podemos melhorar o modelo
# Utilizando GridSearch
# Pegando uma base de dados maior
# Testando outros parâmetros
```

```python
# Fazer novas previsões
# O melhor modelo foi a árvore de decisão com precisão superior ao KNN

# Importar novos clientes
novos_clientes = pd.read_csv("novos_clientes.csv")

novos_clientes["profissao"] = codificador_profissao.transform(novos_clientes["profissao"])
novos_clientes["mix_credito"] = codificador_mix_credito.transform(novos_clientes["mix_credito"])
novos_clientes["comportamento_pagamento"] = codificador_comportamento_pagamento.transform(novos_clientes["comportamento_pagamento"])

previsao = arvore_decisao.predict(novos_clientes)

display(novos_clientes.head(10))
display(previsao)
```
